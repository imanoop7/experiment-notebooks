```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings


embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
def load_data(file_path):
    loader = PyPDFLoader(file_path)
    pages = loader.load_and_split()
    print("loading embeddings in vector db")
    db=FAISS.from_documents(pages,embedding)
    print("saving index")
    db.save_local('faiss')
    return "File Processed"
```

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

torch.random.manual_seed(0)

model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3.5-mini-instruct")
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
)
generation_args = {
    "max_new_tokens": 500,
    "return_full_text": False,
    "temperature": 0.0,
    "do_sample": False,
}
```

```python
def get_response(question,context):
    new_db = FAISS.load_local("faiss", embedding,allow_dangerous_deserialization=True)
    res = new_db.similarity_search(question)
    print(res[0].page_content)
    context = res[0].page_content + res[1].page_content + res[2].page_content
    prompt=f"""
    You are RAG assistant, you will answer the question based on Context.
    Question:{question}
    context:{context}
    """
    response = pipe(prompt, **generation_args)
    print(response)
    return response[0]["generated_text"]


```

```python
import gradio as gr

with gr.Blocks() as demo:
    gr.Markdown("Chatbot")
    with gr.Row():
        inp = gr.File()
        out = gr.Textbox()
    btn = gr.Button("Upload Document")
    btn.click(fn=load_data, inputs=inp, outputs=out)
    with gr.Row():
        inp = gr.Textbox()
        out = gr.Textbox()
    btn = gr.Button("Ask")
    btn.click(fn=get_response, inputs=inp, outputs=out)

demo.launch(debug=True)


```
