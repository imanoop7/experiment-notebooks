```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings


embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
def load_data(file_path):
    loader = PyPDFLoader(file_path)
    pages = loader.load_and_split()
    print("loading embeddings in vector db")
    db=FAISS.from_documents(pages,embedding)
    print("saving index")
    db.save_local('faiss')
    return "File Processed"
```

```python
from langchain_huggingface import HuggingFaceEndpoint,ChatHuggingFace, HuggingFacePipeline
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

tokenizer = AutoTokenizer.from_pretrained("NousResearch/Llama-3.2-1B")
model = AutoModelForCausalLM.from_pretrained("NousResearch/Hermes-3-Llama-3.2-3B",
                                             torch_dtype = torch.float16,
                                             device_map = "auto",
                                             trust_remote_code = True)
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
)
```

```python
llm = HuggingFacePipeline(pipeline=pipe).bind(skip_prompt=True)
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate

def get_response(question):
    new_db = FAISS.load_local("faiss", embedding, allow_dangerous_deserialization=True)
    res = new_db.similarity_search_with_score(question,k=4)
    threshold = 0.7
    filtered_res = [doc for doc, distance in res if distance < threshold]

    if not filtered_res:
        return "No relevant information found."

    context = "\n\n".join([doc.page_content for doc in filtered_res])

    print(context)
    print("-------------------------------------------------------")

    prompt_template = """
    You are an AI-powered assistant. Your task is to answer the given question based on the provided context.

    **Question:** {question}

    **Context:**
    {context}
    """
    prompt = PromptTemplate(
    input_variables=["question","context"], template=prompt_template
    )
    chain = prompt | llm | StrOutputParser()
    response = chain.invoke({"question": question, "context": context})

    return response
```

```python
with gr.Blocks() as demo:
    gr.Markdown("# Chatbot")
    with gr.Row():
        file_input = gr.File()
        file_output = gr.Textbox()
    upload_btn = gr.Button("Upload Document")
    upload_btn.click(fn=load_data, inputs=file_input, outputs=file_output)

    with gr.Row():
        user_input = gr.Textbox(label="Input")
        bot_output = gr.Textbox(label="Output")
    ask_btn = gr.Button("Ask")
    ask_btn.click(fn=get_response, inputs=user_input, outputs=bot_output)

demo.launch(debug=True)
```
